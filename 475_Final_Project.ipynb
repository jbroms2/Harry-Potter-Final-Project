Plotly Visualizations

```python
df_hp_filtered = df_hp[(df_hp['House'] != 'Beauxbatons Academy of Magic') & (df_hp['House'] != 'Durmstrang Institute')]
df_hp_house = df_hp_filtered.groupby('House', as_index = False).agg({'Id': 'count'})
df_hp_house



fig = px.bar(
    df_hp_house, 
    x = 'House', 
    y ='Id', 
    color = 'House', 
    color_discrete_map = {
        'Slytherin' : 'green',
        'Gryffindor': 'maroon',
        'Hufflepuff': 'yellow',
        'Ravenclaw': 'steelblue'
    },
    title='Characters in Each House')

fig.show()

```

```python
df_house_by_gender = df_hp.groupby(['House', 'Gender'], as_index = False).agg({'Name': 'count'}).rename(columns = {'Name': 'Count'})
df_house_by_gender

fig = px.bar(df_house_by_gender, x = 'House', y = 'Count' , color = 'Gender', title = 'House By Gender')
fig.show()
```

```python
df_hp['Species'] = df_hp['Species'].str.strip()
df_hp_species = df_hp.groupby('Species', as_index = False).agg({'Name': 'count'})
df_hp_species.rename(columns = {'Name': 'Total'}, inplace=True ) 

fig = px.pie(df_hp_species, values = 'Total', names= 'Species', title = 'Genetic Makeup')
fig.show()
```
