# The Number of Characters in Each House
The chart below shows the number of characters in each house, excluding those that attended other wizarding schools and were labeled in this way. We created an additional variable to be able to exclude the additional schools in the character list data set.

Then we utilized a groupby function from our filtered dataset to cound how many characters were in each house.

Finally, we created a bar chart in plotly specifically utilizing colors that coordinate with each of the houses.



```python
df_hp_filtered = df_hp[(df_hp['House'] != 'Beauxbatons Academy of Magic') & (df_hp['House'] != 'Durmstrang Institute')]
df_hp_house = df_hp_filtered.groupby('House', as_index = False).agg({'Id': 'count'})
df_hp_house



fig = px.bar(
    df_hp_house, 
    x = 'House', 
    y ='Id', 
    color = 'House', 
    color_discrete_map = {
        'Slytherin' : 'green',
        'Gryffindor': 'maroon',
        'Hufflepuff': 'yellow',
        'Ravenclaw': 'steelblue'
    },
    title='Characters in Each House')

fig.show()
```

# House by Gender Bar Chart
We had to utilize the groupby function in order to group the number of those in each house and the number of each gender that were in each house. We renamed the 'name' column where we had originally aggregated our numbers from to 'count'.

We utilized plotly to make a bar chart where you could see the numbers and colors of each gender separated by each house including the Beauxbatons Academy of Magic and Durmstrang Institute.

```python
df_house_by_gender = df_hp.groupby(['House', 'Gender'], as_index = False).agg({'Name': 'count'}).rename(columns = {'Name': 'Count'})
df_house_by_gender

fig = px.bar(df_house_by_gender, x = 'House', y = 'Count' , color = 'Gender', title = 'House By Gender')
fig.show()
```

# Genetic Makeup of Harry Potter Characters
We created a pie chart of the breakdown of each species. For the Species column, there were additional 'human' categories likely from typing error or capitalization that we first needed to strip down. Then we again used a groupby function to count how many were in each species category and renamed the name column to the total number of those in each species category.

Finally, you will we used a pie chart in plotly to show the breakdown that the vast majority of characters are human.

```python
df_hp['Species'] = df_hp['Species'].str.strip()
df_hp_species = df_hp.groupby('Species', as_index = False).agg({'Name': 'count'})
df_hp_species.rename(columns = {'Name': 'Total'}, inplace=True ) 

fig = px.pie(df_hp_species, values = 'Total', names= 'Species', title = 'Genetic Makeup')
fig.show()
```
